```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
def pars_pdf(text):
    text2 = text.split('\n')    
    print(text2)
    if 'Рейтинг' in text2:
        num = text2.index('Рейтинг')
        raiting = text2[num-1]
        name = text2[num+1]
        activity = text2[num+2]
        textArticle = ''.join(text2[num+4]).replace('\n',' ')
        for i in text2:
            if (re.findall(r'назад', i)):                
                date_publish = i.replace(name, "")
        for i in text2:
            if (re.match(r'Автор оригинала:', i)):
                textArticle = ''.join(text2[num+4]).replace('\n',' ')
                break;
            else:
                textArticle = ''.join(text2[num+3]).replace('\n',' ')
        print(num, raiting, name, activity, date_publish, textArticle)
        
    else:
        raiting = 0
        name = ""
        activity = ""
        if 'Простой' in text2:
            num2 = text2.index('Простой')
            date_publish = text2[num2-1]
        else:
            date_publish = ""
    return  name, raiting, activity, date_publish, textArticle
```

```python
def preprocess_text(data, stopwords=stopwords.words("russian")):
    stopwords.extend(['либо', 'это', 'мб', 'далее', 'дв', 'свой', 'ваш','всё', 'очень', 'её', 'ещё', 'вообще', 'наш', 'который',])
    text = re.sub('ё','е', data.lower())
    text = re.sub(r"([.,!?])", r" \1 ",text)
    text = re.sub(r"[^а-яА-Я\s]+", "",text)
    text = text.strip()
    text = [w for w in text.split() if w not in stopwords] 
    text = [w for w in text if len(w) >=3]
    return ' '.join(text)
```

```python
def lemmatization_text(data, morph=morph):
    result = ' '.join([morph.parse(x)[0].normal_form for x in data.split()])
    return result
```

```python
def get_result(data, morph=morph, stopwords=stopwords.words("russian")):
    result = preprocess_text(data=data)
    result = lemmatization_text(result)
    return result
```

```python
stopword = nltk.corpus.stopwords.words('russian')
```

```python
stopword.extend(['либо', 'это', 'мб', 'далее', 'дв', 'свой', 'ваш','всё', 'очень', 'её', 'ещё', 'вообще', 'наш', 'который', 'также',])    
```

```python
def transform_data(data: pd.Series) -> list:
    result = [get_result(data=i) for i in data]
    return result
```
